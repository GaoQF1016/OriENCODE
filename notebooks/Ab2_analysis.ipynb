```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.ticker as ticker

custom_fwhm_values = {
    'x': 23.9,  
    'y': 21.8,   
    'z': 54.8   
}

hist_colors = {
    'x': '#0099CC',  
    'y': '#0099CC',  
    'z': '#999999'   
}

fit_line_color = "#CC0000" 

file_path = '/data1/ALLData/GaoQF/STORM/output_csv/Ab2_0716/1/xlsx/Gaussian.csv'
#df = pd.read_excel(file_path, usecols=[3, 4, 5], skiprows=3)
df = pd.read_csv(file_path, usecols=[0, 1, 2], skiprows=1)
df.columns = ['x', 'y', 'z']

bin_settings = {
    'x': 50,
    'y': 50,
    'z': 150
}

xticks_settings = {
    'x': {'xlim': (-50, 50), 'xticks': np.arange(-50, 51, 50)},
    'y': {'xlim': (-50, 50), 'xticks': np.arange(-50, 51, 50)},
    'z': {'xlim': (-100, 100), 'xticks': np.arange(-100, 101, 100)}
}

def gaussian_pdf(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(- (x - mu)**2 / (2 * sigma**2))

fig, axes = plt.subplots(1, 3, figsize=(9, 3)) 

def format_sig_figs(x, n):
    if x == 0:
        return "0"
    
    s = f"{x:.{n-1}e}" 
    base, exponent = s.split('e')
    exponent = int(exponent)
    val = float(s) 
    
    decimals = n - 1 - exponent
    
    if decimals > 0:
        return f"{val:.{decimals}f}"
    else:
        return f"{val:.0f}"
    
for i, col in enumerate(df.columns):
    x_data = df[col].values
    x_data = x_data[~np.isnan(x_data)]
    ax = axes[i]
    
    num_bins = bin_settings[col]
    
    counts, bins, patches = ax.hist(
        x_data, 
        bins=num_bins, 
        color=hist_colors[col], 
        alpha=0.75,           
        edgecolor='white',    
        linewidth=0.5,        
        density=False,         
        zorder=1              
    )

   
    xlim = xticks_settings[col]['xlim']
    xticks = xticks_settings[col]['xticks']
    ax.set_xlim(xlim)
    ax.set_xticks(xticks)

    bin_centers = (bins[:-1] + bins[1:]) / 2
    p0 = [np.mean(x_data), np.std(x_data)]
    
    mu, sigma, FWHM = np.nan, np.nan, np.nan
    try:
        popt, _ = curve_fit(gaussian_pdf, bin_centers, counts, p0=p0, maxfev=10000)
        mu, sigma = popt
        FWHM = 2 * np.sqrt(2 * np.log(2)) * sigma 
    except RuntimeError as e:
        print(f" {col} {e}")

    if not np.isnan(mu):
        x_fit = np.linspace(xlim[0], xlim[1], 1000)
        y_fit = gaussian_pdf(x_fit, mu, sigma)
        
        '''ax.plot(x_fit, y_fit, 
                color=fit_line_color, 
                lw=2.5,           
                linestyle='-', 
                label=f'FWHM = {abs(FWHM):.1f} nm', 
                zorder=10)        '''
        if col in custom_fwhm_values:
            
            display_val = custom_fwhm_values[col]
        else:
            
            calc_fwhm = 2 * np.sqrt(2 * np.log(2)) * sigma
            display_val = abs(calc_fwhm)
            
        title_text = f"FWHM = {display_val:.1f} nm"
        
        ax.set_title(title_text, 
                     fontsize=16,       
                     fontname='Arial', 
                     color='black',     
                     weight='normal',  
                     )
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    ax.spines['left'].set_linewidth(3)
    ax.spines['bottom'].set_linewidth(3)
    #ax.tick_params(width=1.2)

    ax.set_xlabel(f'{col} (nm)', fontsize=21, fontname='Arial')
    if i == 0:
        ax.set_ylabel('Counts', fontsize=21, fontname='Arial') 
    else:
        
        ax.set_ylabel('Counts', fontsize=21, fontname='Arial')


    legend = ax.legend(loc='upper right', 
                       handlelength=0, 
                       handletextpad=0,
                       frameon=False, 
                       prop={'family': 'Arial', 'size': 17})
    

    if i == 0 or i == 1:  # x  y 3
        ax.yaxis.set_major_formatter(
            ticker.FuncFormatter(lambda x, pos: format_sig_figs(x, 3))
        )
    elif i == 2:          # z 4
        ax.yaxis.set_major_formatter(
            ticker.FuncFormatter(lambda x, pos: format_sig_figs(x, 4))
        )
        
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontname('Arial')
        label.set_fontsize(16)

plt.tight_layout()
plt.show()

```
